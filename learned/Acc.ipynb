```python
import json

# 加载 JSON 文件
def load_json(file_path):
    with open(file_path, "r") as file:
        return json.load(file)

# 统计预测结果是否包含异常
def analyze_predictions(results):
    # 提取预测标签
    predicted_labels = [
        "structural anomaly" if "Yes" in item["predicted_answer"] else "normal"
        for item in results
    ]
    
    # 统计总样本数和预测为异常的样本数
    total_samples = len(predicted_labels)
    anomaly_predictions = predicted_labels.count("structural anomaly")
    
    return {
        "total_samples": total_samples,          # 总样本数
        "anomaly_predictions": anomaly_predictions,  # 预测为异常的样本数
        "normal_predictions": total_samples - anomaly_predictions  # 预测为正常的样本数
    }

# 主函数
if __name__ == "__main__":
    # 文件路径
    file_path = "../../../../result/juice-bottle/output/result_strcutural.json"
    
    # 加载数据
    results = load_json(file_path)
    
    # 分析预测结果
    prediction_analysis = analyze_predictions(results)
    
    # 输出结果
    print("分析结果：")
    print(f"总样本数: {prediction_analysis['total_samples']}")
    print(f"预测为异常的样本数: {prediction_analysis['anomaly_predictions']}")
    print(f"预测为正常的样本数: {prediction_analysis['normal_predictions']}")
```
